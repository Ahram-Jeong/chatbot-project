```python
import requests as req
import json
from tqdm import tqdm
```

```python
# 1. ì¹´ë“œ ë¦¬ìŠ¤íŠ¸ ìˆ˜ì§‘ (ì‹ ìš© ì¹´ë“œ)
page = 1
all_cards = []

while True :
    url = f"https://api.card-gorilla.com:8080/v1/cards?p={page}&perPage=10&cate=CRD"
    res = req.get(url)
    if res.status_code != 200 :
        print(f"ì‘ë‹µ ì‹¤íŒ¨! status code: {res.status_code}")
        break
    
    data = res.json() # ì‘ë‹µë°›ì€ ë°ì´í„°ë¥¼ JSONìœ¼ë¡œ
    
    # ë°ì´í„°ê°€ ë¹„ì–´ìˆìœ¼ë©´ ì¢…ë£Œ
    if not data["data"] :
        break
    
    all_cards.extend(data["data"])
    print(f"Page {page} ì™„ë£Œ")
    page += 1 # ë‹¤ìŒ í˜ì´ì§€
    
# JSON ì •ë³´ë¥¼ íŒŒì¼ë¡œ ì €ì¥
with open("../data/all_credit.json", "w", encoding = "utf-8") as f :
    json.dump(all_cards, f, ensure_ascii = False, indent = 4)
print("ì™„ë£ŒğŸ˜‡")
```

```python
# 1. ì¹´ë“œ ë¦¬ìŠ¤íŠ¸ ìˆ˜ì§‘ (ì²´í¬ ì¹´ë“œ)
page = 1
all_cards = []

while True :
    url = f"https://api.card-gorilla.com:8080/v1/cards?p={page}&perPage=10&cate=CHK"
    res = req.get(url)
    if res.status_code != 200 :
        print(f"ì‘ë‹µ ì‹¤íŒ¨! status code: {res.status_code}")
        break
    
    data = res.json()  # ì‘ë‹µë°›ì€ ë°ì´í„°ë¥¼ JSONìœ¼ë¡œ
    
    # ë°ì´í„°ê°€ ë¹„ì–´ìˆìœ¼ë©´ ì¢…ë£Œ
    if not data["data"] :
        break
    
    all_cards.extend(data["data"])
    print(f"Page {page} ì™„ë£Œ")
    page += 1 # ë‹¤ìŒ í˜ì´ì§€
    
# JSON ì •ë³´ë¥¼ íŒŒì¼ë¡œ ì €ì¥
with open("../data/all_chk.json", "w", encoding = "utf-8") as f :
    json.dump(all_cards, f, ensure_ascii = False, indent = 4)
print("ì™„ë£ŒğŸ˜‡")
```

```python
# 2. ì¹´ë“œ ìƒì„¸ì •ë³´ ìˆ˜ì§‘ (ì‹ ìš© ì¹´ë“œ)
input_path = "../data/all_credit.json"
output_path = "../data/detail_credit.json"

# JSON íŒŒì¼ ì½ê¸°
with open(input_path, "r", encoding = "utf-8") as f :
    data = json.load(f)
    
# "cid" ê°’ì„ ì‚¬ìš©í•˜ì—¬ API í˜¸ì¶œ í›„ ê²°ê³¼ ì €ì¥
result_data = []

for i in tqdm(data) :
    cid = i.get("cid")
    if cid :
        api_url = f"https://api.card-gorilla.com:8080/v1/cards/{cid}"
        res = req.get(api_url)
        if res.status_code == 200 :
            result_data.append(res.json())
        else :
            print(f"ì‹¤íŒ¨ cid : {cid}")
# ê²°ê³¼ ë°ì´í„°ë¥¼ JSON íŒŒì¼ë¡œ ì €ì¥
with open(output_path, "w", encoding = "utf-8") as f :
    json.dump(result_data, f, ensure_ascii = False, indent = 4)
```
